# Método de Riemann

```julia
function Riemann(f,a,b,n)
    list = linspace(a,b,n)
    I = 0
    for i in 1:length(list)-1
        a = list[i]
        b = list[i+1]
        I += (b-a)*f((a+b)/2)
    end
    return I
end;
```

```julia
function Trapecio(f,a,b,n)
    list = linspace(a,b,n)
    I = 0
    for i in 1:length(list)-1
        a = list[i]
        b = list[i+1]
        I += (b-a)*(f(a)+f(b))/2
    end
    return I
end
```

```julia
function Simpson(f,a,b,n)
    list = linspace(a,b,n)
    I = 0
    for i in 1:length(list)-1
        a = list[i]
        b = list[i+1]
        I += (b-a)*(f(a)+4*f((a+b)/2)+f(b))/6
    end
    return I
end
```

```julia
f(x)=10*x^4+2;
```

```julia
Riemann(f,0,2*pi,2)
```

```julia
exact=4*(π+16*π^5);
```

```julia
function convergence_check(f,method,a,b,exact)
    list=[];
       for i in 2:10
        push!(list,method(f,a,b,i))
    end
    return abs.((list-exact)/exact);
end;
```

```julia
list1 = convergence_check(f,Riemann,0,2*pi,exact);
list2 = convergence_check(f,Trapecio,0,2*pi,exact);
list3 = convergence_check(f,Simpson,0,2*pi,exact);
```
