# MÃ©todo de Riemann

```julia
function Riemann(f,a,b,n)
    list = linspace(a,b,n)
    I = 0
    for i in 1:length(list)-1
        a = list[i]
        b = list[i+1]
        I += (b-a)*f((a+b)/2)
    end
    return I
end
```

```julia
Riemann(x->x^2,1,3,100)
```

```julia
function Trapecio(f,a,b,n)
    list = linspace(a,b,n)
    I = 0
    for i in 1:length(list)-1
        a = list[i]
        b = list[i+1]
        I += (b-a)*(f(a)+f(b))/2
    end
    return I
end
```

```julia
Trapecio(x->x^2,1,3,100)
```

```julia
function Simpson(f,a,b,n)
    list = linspace(a,b,n)
    I = 0
    for i in 1:length(list)-1
        a = list[i]
        b = list[i+1]
        I += (b-a)*(f(a)+4*f((a+b)/2)+f(b))/6
    end
    return I
end
```

```julia
Simpson(x->x^2,1,3,100)
```

```julia
division = range(5,5,4)
```

```julia
list1 = zeros(length(division))
for i in 1:length(division)
    list1[i] = Riemann(x->cos.(x)+1,-pi,pi,division[i])
end
```

```julia
error1 = abs.((list1-2*pi)/2*pi);
```

```julia
list2 = zeros(length(division))
for i in 1:length(division)
    list2[i] = Trapecio(x->cos.(x)+1,-pi,pi,division[i])
end
```

```julia
error2 = abs.((list2-2*pi)/2*pi);
```

```julia
list3 = zeros(length(division))
for i in 1:length(division)
    list3[i] = Simpson(x->cos.(x)+1,-pi,pi,division[i])
end
```

```julia
error3 = abs.((list3-2*pi)/2*pi);
```

```julia
using PyPlot
```

```julia
plot(error1,label="Riemann")
plot(error2,label="Trapecio")
plot(error3,label="Simpson")
legend()
```
