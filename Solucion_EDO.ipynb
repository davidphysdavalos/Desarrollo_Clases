```julia
function ord_diff_eqn_solve1(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx = zeros(length(list))
     listx[1] = x0
     for i in 2:length(list)
        t = i*h
        x = x + f(x,t)
        listx[i] = x
     end
     return listx, list
end
```

```julia
f(x,t) = -2.5*x
```

```julia
listt = range(1,1,8)
```

```julia
last(listt)
```

```julia
(x,t) = ord_diff_eqn_solve1(f,listt,1)
```

```julia
plot(t,x)
```

```julia
(x2,t2) = ord_diff_eqn_solve1(f,listt,2);
```

```julia
plot(t2,x2)
```

```julia
function ord_diff_eqn_solve2(f,list,x0)
     x = x0
     h = (last(list)-first(list))/length(list)
     listx = zeros(length(list))
     for i in 1:length(list)
        t = list[i]
        x = x + f(x,t)
        t2 = t + h
        for j in 1:20 
            x = x + f(x,t2)*h
        end
        listx[i] = x
     end
     return listx, list
end
```

```julia
(x3,t3) = ord_diff_eqn_solve2(f,listt,1)
```

```julia
plot(t3,x3)
```
